```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
plt.scatter(df['R&D Spend'],df['Profit'],color = 'Blue', marker ='o')
plt.xlabel("R&D Spend")
plt.ylabel("Profit")
plt.title("R&D vs Profit")
plt.show()
```

```python
x=df[["R&D Spend"]]
y=df["Profit"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42) 
```

```python
Lr = LinearRegression()         
Lr.fit(x_train, y_train)        
y_pred = Lr.predict(x_test)
```

```python
r2_score=r2_score(y_test,y_pred)
MSE=mean_squared_error(y_test,y_pred) 
```

```python
import gradio as gr
import joblib
import numpy as np

# Load your trained SVM model
model = joblib.load(r"C:\\Users\\Hp\\OneDrive\Desktop\\ML\\startup.pkl")

# Define the prediction function
def predict_profit(rd_spend):
    features = np.array([rd_spend])
    prediction = model.predict(features)
    return f"Predicted profit: {prediction[0]}"

# Define Gradio interface
inputs = [
    gr.Number(label="YearsExperience")

   ]

output = gr.Textbox(label="Prediction")
gr.Interface(fn=predict_profit, inputs=inputs, outputs=output, title="startup predictor").launch()
```
